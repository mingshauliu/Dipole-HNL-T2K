```python
### let's calculate rought HNL upscattering cross section in the lead shielding at CCM from
### https://arxiv.org/pdf/1803.03262.pdf


def rough_xsect(d, m_HNL, Z, rad):
    d, m_HNL, Z, rad = d, m_HNL, Z, rad
    alpha = 1/137
    E_nu = 0.03 ## nu energy in gev
    hc = 1.973e-16 ## hbar*c in gev*m

    sigma_gev = 4 * alpha * Z**2 * d**2 * np.log((4*E_nu**2) / (m_HNL**4 * (rad/hc)**2)) ## units of gev^-2
    sigma_cm = (sigma_gev*hc**2)*100**2
    return sigma_cm

d_6 = 10e-6 ## 1/gev 
d_7 = 10e-7 ## 1/gev
m_HNL = np.linspace(0.01, 0.03, 10) ## mass of HNL in GeV (or 1 to 30 mev)
Z_pb = 82 ## lead Z number
rad_pb = 5.56e-15 ## radius of lead in m

Z_fe = 26 ## lead Z number
rad_fe = 4e-15 ## radius of iron in m


fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(m_HNL*1000, rough_xsect(d_6, m_HNL, Z_pb, rad_pb), 'b-', label = "Pb, d = 10e-6 1/gev")
ax.plot(m_HNL*1000, rough_xsect(d_7, m_HNL, Z_pb, rad_pb),'b--', label = "Pb, d = 10e-7 1/gev")
ax.plot(m_HNL*1000, rough_xsect(d_6, m_HNL, Z_fe, rad_fe), 'r-', label = "Fe, d = 10e-6 1/gev")
ax.plot(m_HNL*1000, rough_xsect(d_7, m_HNL, Z_fe, rad_fe), 'r--', label = "Fe, d = 10e-7 1/gev")
ax.set_title("Approximate Cross Section for HNL Upscattering", fontsize = 16)
ax.set_xlabel("Mass of HNL (MeV)", fontsize = 16)
ax.set_ylabel("Cross Section (cm$^2$)", fontsize = 16)
ax.legend(fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.semilogy()
plt.grid()
plt.xlim(0, 30)
# plt.ylim(10e-38, 10e-33)


```

```python
## now let's see how many events we predict to see in one pulse

### cross section (cm^2/lead target) * nu_flux (nu/(cm^2*s)) * number of lead targets * data collection time

m_HNL = np.linspace(0.001, 0.03, 100) ## mass of HNL in GeV (or 1 to 30 mev)

pb6 = rough_xsect(d_6, m_HNL, Z_pb, rad_pb) ##cm^2
pb7 = rough_xsect(d_7, m_HNL, Z_pb, rad_pb) ##cm^2
fe6 = rough_xsect(d_6, m_HNL, Z_fe, rad_fe) ##cm^2
fe7 = rough_xsect(d_7, m_HNL, Z_fe, rad_fe) ##cm^2


nu_flux = 4.74e5 ## nu/ cm^2/s
# data_time = 9.467e+7 # seconds in 3 years
data_time = 15780000 # seconds in 6 months
#data_time = 604800 ##seconds in a week

nuclei_gram_fe = 6.0221408e+23/55.845 ## nuclei/mole * mole/gram of fe = nuclei/gram of fe
fe_mass = 100*258*225*7.9 ## weight of steel shielding in g
# target_fe = nuclei_gram_fe * fe_mass ## number of steel nuclei targets

target_shielding_vol = (6.44/360) * 236870506 ## vol of small part of donut in cm^3
target_shielding_mass = target_shielding_vol*7.9 ## weight of steel shielding in g
target_fe = (nuclei_gram_fe * fe_mass) + (nuclei_gram_fe * target_shielding_mass) ## number of steel nuclei targets

nuclei_gram_pb = 6.0221408e+23/207.2 ## nuclei/mole * mole/gram of pb = nuclei/gram of pb
pb_mass = 30*258*225*11.34 ## weight of pb shielding in g
target_pb = nuclei_gram_pb* pb_mass


solid_ang = (2.58*2.25)/(4*np.pi)

events_targets_6_fe = fe6*nu_flux*target_fe*data_time*solid_ang
events_targets_7_fe = fe7*nu_flux*target_fe*data_time*solid_ang
events_targets_6_pb = pb6*nu_flux*target_pb*data_time*solid_ang
events_targets_7_pb = pb7*nu_flux*target_pb*data_time*solid_ang

##5 m of steel around target!! need to account for that, 6months of data  

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(m_HNL*1000, events_targets_6_fe,'r-', label = "Steel, d = 10e-6 1/gev")
ax.plot(m_HNL*1000, events_targets_7_fe, 'r--', label = "Steel, d = 10e-7 1/gev")
ax.plot(m_HNL*1000, events_targets_6_fe + events_targets_6_pb, 'b-', label = "Steel + 30 cm Lead, d = 10e-6 1/gev")
ax.plot(m_HNL*1000, events_targets_7_fe + events_targets_7_pb, 'b--', label = "Steel + 30 cm Lead, d = 10e-7 1/gev")
# ax.plot(m_HNL*1000, events_targets_6_pb, 'b-', label = "30 cm Lead, d = 10e-6 1/gev")
# ax.plot(m_HNL*1000, events_targets_7_pb, 'b--', label = "30 cm Lead, d = 10e-7 1/gev")

ax.set_title("HNL events for 6 months of data collection", fontsize = 16)
ax.set_xlabel("Mass of HNL (MeV)", fontsize = 16)
ax.set_ylabel("Number of Events", fontsize = 16)
ax.legend(fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.semilogy()
plt.grid()
plt.xlim(0, 30)
plt.ylim(10e3, 10e8)

```

```python
## prob of HNL decay in CCM

mtau = 1.776 #GeV
mmu = 0.106 #GeV

def length(d):
    dtau = (mtau/mmu)*d
    d = np.sqrt(d**2 + dtau**2)
    E_HNL = 0.03 ## HNL energy in gev
    m_HNL = np.linspace(0.001, 0.03, 100) ## mass of HNL in GeV (or 1 to 30 mev)
    hc = 1.973e-16 ## hbar*c in gev*m
    L = ((4*np.pi*E_HNL*hc)/(m_HNL**4 * d**2)) * np.sqrt(1 - (m_HNL**2/E_HNL**2))
    return L
    
d_6 = 10e-6 ## 1/gev 
d_7 = 10e-7 ## 1/gev

CCM_diameter = 2.58 ## meters

fig, ax = plt.subplots(figsize=(12, 8))
ax.plot( m_HNL, CCM_diameter/length(d_6), 'r-', label = "d = 10e-6 1/gev")
ax.plot( m_HNL, CCM_diameter/length(d_7), 'b-', label = "d = 10e-7 1/gev")
ax.set_title("HNL Decay Probability", fontsize = 16)
ax.set_xlabel("HNL Mass (GeV)", fontsize = 16)
ax.set_ylabel("Decay Probability (%)", fontsize = 16)
ax.legend(fontsize = 16)
# ax.set_xlim(0, 2.25)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.semilogy()
```

```python
### decay time

def decay_time(d):
    d = d ## distance to center of shielding in meter
    E_HNL = 0.03 ## HNL energy in gev
    m_HNL = np.linspace(0.001, 0.03, 100) ## mass of HNL in GeV (or 1 to 30 mev)
    c = 2.99e8 ## speed of light in m/s
    beta = np.sqrt(1 - (m_HNL**2/E_HNL**2)) ## beta in lab frame
    t = d/c + (np.abs((d-20))/(beta*c)) - 20/c ## time in seconds
    t_nano = t*1e+9 ## time in nanosec
    return t_nano


fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(m_HNL*1000, decay_time(d=3),'r-', label = "Steel shielding around target")
ax.plot(m_HNL*1000, decay_time(d=17.5),'b-', label = "Steel shielding around CCM")
ax.axhline(190, color ='black', lw = 2, label = '190ns data collection time')
ax.set_title("Average HNL Decay Time", fontsize = 16)
ax.set_xlabel("Mass of HNL (MeV)", fontsize = 16)
ax.set_ylabel("Time after prompt $\\nu_{\\mu}$ arrival (ns)", fontsize = 16)
ax.legend(fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
# plt.semilogy()
plt.grid()
plt.xlim(0, 30)
# plt.ylim(10e3, 10e8)





```

```python
### let's work on calculating the energy spectrum for the emitted gamma ray in the detector 

def mass_to_events(m_HNL, d):
    m_HNL, d = m_HNL, d
    
    E_HNL = 0.03 ## HNL energy in gev
    hc = 1.973e-16 ## hbar*c in gev*m
    CCM_diameter = 2.58 ## meters
    Z_fe = 26 ## lead Z number
    rad_fe = 4e-15 ## radius of iron in m
#     nu_flux = 4.74e5 ## nu/ cm^2/s !!!old number!!
    nu_flux = 5.28e5 ## nu/ cm^2/s
#     data_time = 15780000 # seconds in 6 months
    data_time = 9.467e+7/2 # seconds in 3 years (1.5 yrs of beam data)
    solid_ang = (2.58*2.25)/(4*np.pi)

    ## 1m of steel shielding in front of detector
    nuclei_gram_fe = 6.0221408e+23/55.845 ## nuclei/mole * mole/gram of fe = nuclei/gram of fe
    fe_mass = 100*258*225*7.9 ## weight of 1m of steel shielding in g
    
    ## 4m radially of steel shielding in front of the target
    target_shielding_vol = (6.44/360) * 236870506 ## vol of small part of donut in cm^3
    target_shielding_mass = target_shielding_vol*7.9 ## weight of steel shielding in g
    target_fe_far = (nuclei_gram_fe * target_shielding_mass) ## number of steel nuclei targets
    target_fe_near = (nuclei_gram_fe * fe_mass) 
    xsect_fe = rough_xsect(d, m_HNL, Z_fe, rad_fe)
    
    ### 2 meters of concrete
    
    ## what is concrete made of? Here is a breakdown to concrete by weight
    
    frxn_by_weight = [0.01, 0.001, 0.5291, 0.016, 0.002, 0.033872, 0.337021, 0.013, 0.044, 0.0124]
    atomic_num = [1, 6, 8, 11, 12, 13, 14, 19, 20, 26]
    charge_rad = [6.65e-15, 2.5e-15, 2.8e-15, 2.9928e-15, 3.0284e-15, 3.0610e-15, 3.1176e-15, 3.4556e-15, 3.4953e-15, 3.7532e-15]
    mol_mass = [1.008, 12.011, 15.999, 22.99, 24.31, 26.982, 28.086, 39.098, 40.074, 55.854]
    avo = [6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23, 6.0221408e+23]
    nuclei_gram_con = [f * i / j for f,i, j in zip(frxn_by_weight,avo, mol_mass)] ## nuclei/mole * mole/gram of conc = nuclei/gram for each element in conc

    con_mass = 200*258*225*2.3 ## weight of 2m of concrete shielding in g
    target_con = con_mass* np.asarray(nuclei_gram_con) ## number of concrete nuclei targets

    xsect_con = [rough_xsect(d, m_HNL, k, l) for k, l in zip(atomic_num, charge_rad)]

    xsect_fe = rough_xsect(d, m_HNL, Z_fe, rad_fe)
    
    ## let's add in the argon in the detector
    
    Z_ar = 18
    rad_ar = 3.4274e-15
    ar_mass = 5e6 ## 5 tons in g
    avo = 6.0221408e+23
    ar_mol_mass = 39.948 ##g/mol
    ar_targets = ar_mass * avo/ar_mol_mass
    
    
    xsect_ar = rough_xsect(d, m_HNL, Z_ar, rad_ar)

    ### let's calculate the probability of decay in detector
    
    dtau = (mtau/mmu)*d
    d = np.sqrt(d**2 + dtau**2)
    

    L = ((4*np.pi*E_HNL*hc)/(m_HNL**4 * d**2)) * np.sqrt(1 - (m_HNL**2/E_HNL**2))
    
#     P = np.exp(-l_ent / L) - np.exp(-l_ex / L)

    events_fe_far = (np.exp(-16 / L) - np.exp(-18 / L)) * (xsect_fe*nu_flux*target_fe_far*data_time*solid_ang)
    events_fe_near = (np.exp(-3 / L) - np.exp(-5 / L)) * (xsect_fe*nu_flux*target_fe_near*data_time*solid_ang)
    events_con = [(np.exp(-1.5 / L) - np.exp(-3.5 / L)) * (x*nu_flux*t*data_time*solid_ang) for x, t in zip(xsect_con, target_con)]
    events_ar = (np.exp(-0 / L) - np.exp(- (CCM_diameter/2)/ L)) * (xsect_ar*nu_flux*ar_targets*data_time*solid_ang)

    events = events_fe_far + events_fe_near + np.sum(events_con) + events_ar
    
    return events, L 


### energy spectrum for majorana case and one for the dirac case

### let's start w majorana
E_nu = 0.03 ## nu energy in gev
m_HNL = np.linspace(0.001, 0.03, 100) ## mass of HNL in GeV (or 1 to 30 mev)
P_HNL = np.sqrt(E_nu**2 - m_HNL**2) ## momentum of HNL

maj_avg_E = E_nu/2
maj_sig_E = np.sqrt(P_HNL**2/12)

dirac_avg_E = E_nu/2 - P_HNL/6
dirac_sig_E = np.sqrt(P_HNL**2/18)


fig, ax = plt.subplots(figsize=(12, 8))
# ax.plot(maj_avg_E*1000+maj_sig_E*1000, mass_to_events(np.linspace(0.001, 0.03, 100), 10e-6),'r-', label = "Avg E (Majorana Case)")
# ax.plot(maj_avg_E*1000-maj_sig_E*1000, mass_to_events(np.linspace(0.001, 0.03, 100), 10e-6),'r-', label = "Avg E (Majorana Case)")
ax.plot(maj_avg_E*1000+maj_sig_E*1000, m_HNL*1000,'r--', label = "1$\\sigma$")
ax.plot(maj_avg_E*1000-maj_sig_E*1000, m_HNL*1000,'r--')

ax.axvline(maj_avg_E*1000, color = 'red', label = "Average Energy (Majorana Case)")

ax.plot(dirac_avg_E*1000+dirac_sig_E*1000, m_HNL*1000,'b--', label = "1$\\sigma$")
ax.plot(dirac_avg_E*1000-dirac_sig_E*1000, m_HNL*1000,'b--')

ax.plot(dirac_avg_E*1000, m_HNL*1000,'b-', label = "Average Energy (Dirac Case)")


ax.set_title("Energy Spectrum of HNL to 1$\\sigma$", fontsize = 16)
ax.set_ylabel("mass", fontsize = 16)
ax.set_xlabel("Average Energy (MeV)", fontsize = 16)
ax.legend(fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.grid()
# plt.semilogy()
# plt.xlim(0, 30)
# plt.ylim(0, 100)
```

```python
### making a contour graph

## defining sigma levels 
sigmas = np.array([1, 2 ])
dof = 2
proportions = scipy.special.erf(sigmas / np.sqrt(2.0))
levels = scipy.special.gammaincinv(dof / 2.0, np.array(proportions))


low_bkgd = 30
high_bkgd = 300

one_sig_low = np.sqrt(low_bkgd)
two_sig_low = 2*np.sqrt(low_bkgd)
one_sig_high = np.sqrt(high_bkgd)
two_sig_high = 2*np.sqrt(high_bkgd)


possible_d = np.logspace(-8, -5, 1000)
m_HNL = np.linspace(0.001, 0.0298, 100)

one_sig_low_m = []
one_sig_low_d = []
one_sig_high_m = []
one_sig_high_d = []

two_sig_low_m = []
two_sig_low_d = []
two_sig_high_m = []
two_sig_high_d = []



best_d_low_two = []
best_d_high_two = []

for i in range(len(m_HNL)):
    print(i)
    for j in range(len(possible_d)):
        events = mass_to_events(m_HNL[i], possible_d[j])[0]
        #print(m_HNL[i],possible_d[j],events)
        
        
        if events > one_sig_low:
            one_sig_low_m.append(m_HNL[i]*1000)
            one_sig_low_d.append(possible_d[j])

        if events > one_sig_high:
            one_sig_high_m.append(m_HNL[i]*1000)
            one_sig_high_d.append(possible_d[j])

        if events > two_sig_low:
            two_sig_low_m.append(m_HNL[i]*1000)
            two_sig_low_d.append(possible_d[j])
            

        if events > two_sig_high:
            two_sig_high_m.append(m_HNL[i]*1000)
            two_sig_high_d.append(possible_d[j])
            
#         if (m_HNL[i] > 0.029 and possible_d[j] < 1.2e-7):
#             print('mass = %s' %m_HNL[i])
#             print('d = %s' %possible_d[j])
#             print('L = %s' % mass_to_events(m_HNL[i], possible_d[j])[1])
#             print('probability = %s' % mass_to_events(m_HNL[i], possible_d[j])[2])
#             print('events = %s' % mass_to_events(m_HNL[i], possible_d[j])[0])
#             print()

            
    best_d_low_two.append(np.min(two_sig_low_d))
    best_d_high_two.append(np.min(two_sig_high_d))
    one_sig_low_m = []
    one_sig_low_d = []
    one_sig_high_m = []
    one_sig_high_d = []

    two_sig_low_m = []
    two_sig_low_d = []
    two_sig_high_m = []
    two_sig_high_d = []
    
    

fig, ax = plt.subplots(figsize=(12, 8))

# ax.plot(one_sig_low_m, one_sig_low_d,'r.', label = "1$\\sigma$ Low Bkgnd")
ax.plot(m_HNL*1000, best_d_low_two,'m-', label = "2$\\sigma$ Low Bkgnd")

# ax.plot(one_sig_high_m, one_sig_high_d, 'g.', label = "1$\\sigma$ High Bkgnd")
ax.plot(m_HNL*1000, best_d_high_two, 'g-', label = "2$\\sigma$ High Bkgnd")

fout_low = open('ConstraintFiles/CCM/lowbkg_constraint_dtau.txt','w')
fout_high = open('ConstraintFiles/CCM/highbkg_constraint_dtau.txt','w')
for m,dlow,dhigh in zip(m_HNL*1000,best_d_low_two,best_d_high_two):
    if m > 29: dlow,dhigh = 1,1
    print(m,dlow,file=fout_low)
    print(m,dhigh,file=fout_high)
fout_low.close()
fout_high.close()

ax.set_title("CCM HNL Sensitivity: 3 Years of Data", fontsize = 16)
ax.set_xlabel("Mass of HNL (MeV)", fontsize = 16)
ax.set_ylabel("d (GeV$^{-1}$)", fontsize = 16)
ax.legend(fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.semilogy()
plt.semilogx()
plt.grid()
plt.xlim(10, 10e3)
plt.ylim(10e-9, 10e-5)

```
